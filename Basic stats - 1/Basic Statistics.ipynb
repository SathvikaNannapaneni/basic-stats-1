```python
import pandas as pd

# Load the dataset
df = pd.read_csv("C:\\Users\\sathv\\Downloads\Basic stats - 1\\Basic stats - 1\\sales_data_with_discounts.csv")

# Identify numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

# Calculate and display statistics for each numerical column
for col in numerical_cols:
    mean = df[col].mean()
    median = df[col].median()
    mode = df[col].mode().values[0]  # mode() returns a Series, so we take the first value
    std_dev = df[col].std()

    print(f"Column: {col}")
    print(f"  Mean: {mean}")
    print(f"  Median: {median}")
    print(f"  Mode: {mode}")
    print(f"  Standard Deviation: {std_dev}")
    print()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("C:\\Users\\sathv\\Downloads\Basic stats - 1\\Basic stats - 1\\sales_data_with_discounts.csv")

# Identify numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

# Create histograms and boxplots
for col in numerical_cols:
    plt.figure(figsize=(14, 6))

    # Histogram
    plt.subplot(1, 2, 1)
    plt.hist(df[col].dropna(), bins=30, edgecolor='k', alpha=0.7)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')

    plt.tight_layout()
    plt.show()
from scipy.stats import skew, kurtosis

# Analyze distribution for each numerical column
for col in numerical_cols:
    data = df[col].dropna()
    
    # Skewness
    skewness = skew(data)
    
    # Kurtosis
    kurt = kurtosis(data)
    
    print(f"Column: {col}")
    print(f"  Skewness: {skewness:.2f}")
    print(f"  Kurtosis: {kurt:.2f}")
    print()
```

```python
#●Bar Chart Analysis for Categorical Column
import pandas as pd

# Load the dataset
df = pd.read_csv("C:\\Users\\sathv\\Downloads\Basic stats - 1\\Basic stats - 1\\sales_data_with_discounts.csv")

# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

print("Categorical columns:")
print(categorical_cols)

import seaborn as sns
import matplotlib.pyplot as plt

# Create bar charts for each categorical column
for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```

```python
#Standardization of Numerical Variables

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("C:\\Users\\sathv\\Downloads\Basic stats - 1\\Basic stats - 1\\sales_data_with_discounts.csv")

# Identify numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

# Create a copy of the DataFrame to avoid modifying the original
df_standardized = df.copy()

# Standardize the numerical columns
for col in numerical_cols:
    mu = df[col].mean()
    sigma = df[col].std()
    df_standardized[col] = (df[col] - mu) / sigma

# Plot histograms for original and standardized columns
for col in numerical_cols:
    plt.figure(figsize=(14, 6))

    # Original Data
    plt.subplot(1, 2, 1)
    plt.hist(df[col].dropna(), bins=30, edgecolor='k', alpha=0.7)
    plt.title(f'Original Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

    # Standardized Data
    plt.subplot(1, 2, 2)
    plt.hist(df_standardized[col].dropna(), bins=30, edgecolor='k', alpha=0.7)
    plt.title(f'Standardized Distribution of {col}')
    plt.xlabel(f'Standardized {col}')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()
```
